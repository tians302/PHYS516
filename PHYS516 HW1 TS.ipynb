```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

all_x = []
all_y = []
all_z = []
all_theta = []
all_phi = []

# Generate points using the transformation
for n in range(1000):
    u = random.random()
    v = random.random()
    theta = np.arccos(1 - 2*u) 
    phi = 2*np.pi*v
    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)
    
    all_theta.append(theta)
    all_phi.append(phi)
    all_x.append(x)
    all_y.append(y)
    all_z.append(z)
    # print(x,y,z)


fig1 = plt.figure()
ax = Axes3D(fig1)
ax.scatter(all_x, all_y, all_z)
plt.title('3D plot')
plt.grid(True)
plt.show()

fig2 = plt.figure()
plt.scatter(all_phi, all_theta)
plt.xlabel('phi')
plt.ylabel('theta')
plt.title('2D plot')
plt.grid(True)
plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

all_x = []
all_y = []
all_z = []

# Generate points using the rejection method
for i in range(1000):
    while True:
        a = 2*random.random() - 1
        b = 2*random.random() - 1
        c = 2*random.random() - 1
        l = a**2 + b**2 + c**2 # squared length
        if l <= 1: #check if it is inside the sphere
            break
    norm = np.sqrt(l) #normalized, projection onto surface
    x = a / norm
    y = b / norm
    z = c / norm
    all_x.append(x)
    all_y.append(y)
    all_z.append(z)

fig1 = plt.figure()
ax = Axes3D(fig1)
ax.scatter(all_x, all_y, all_z)
plt.title('3D plot using Rejection Method')
plt.grid(True)
plt.show()

fig2 = plt.figure()
plt.scatter(all_phi, all_theta)
plt.xlabel('phi')
plt.ylabel('theta')
plt.title('2D plot using Rejection Method')
plt.grid(True)
plt.show()
```
