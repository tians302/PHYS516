```python
import math

# gold section routine
def gold_section(f, a, b):
    r = (math.sqrt(5) - 1) / 2  # gold ratio
    tol = 0.00001 # tolerance

    c = b - (b - a) * r
    d = a + (b - a) * r

    while abs(b - a) > tol:
        if f(c) < f(d):
            b = d
        else:
            a = c

        # update c & d
        c = b - (b - a) * r
        d = a + (b - a) * r
        
        xmin = (b + a)/2
    return xmin

# function1
def f1(x):
    return math.sin(math.exp(x))

x1_min = gold_section(f1, 1, 2)

print('minimal x for f1 is:', x1_min, ', min f1 is:', f1(x1_min))
```

```python
def f2(x):
    if x != 0:
        return x**x
    else:
        return 1  # special case

x2_min = gold_section(f2, 0, 1)
print('minimal x for f2 is:', x2_min, ', min f2 is:', f2(x2_min))
```

```python
# plot to verify
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,2,100)
y = [f1(x) for x in x]
plt.figure(figsize=(10,8))
plt.plot(x, y)
plt.title('f1(x) = sin(e^x)')
plt.xlabel('x')
plt.ylabel('f1(x)')
plt.show

# we can see min is around 1.5
```

```python
x = np.linspace(0,1,100)
y = [f2(x) for x in x]
plt.figure(figsize=(10,8))
plt.plot(x, y)
plt.title('f2(x) = x^x')
plt.xlabel('x')
plt.ylabel('f2(x)')
plt.show
# we can see min is around 0.36
```
